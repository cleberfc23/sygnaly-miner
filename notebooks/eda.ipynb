```python
import pandas as pd
import sys
import os
import numpy as np 
import warnings
warnings.filterwarnings("ignore")
```

```python
data_frame_full = pd.read_csv("hf://datasets/WiktorS/polish-news/full_data.csv")
```

```python
size = 1000
data_frame = data_frame_full[0:size]
data_frame.head()
```

```python
print(data_frame.shape)
print(data_frame.columns)
```

```python
for i in range(0, 2):
    print(f"Link: {data_frame['link'][i]}")
    print(f"Title ({len(data_frame['title'][i])}): {data_frame['title'][i]}")
    print(
        f"Headline ({len(data_frame['headline'][i])}): {data_frame['headline'][i]}")
    print(
        f"Content ({len(data_frame['content'][i])}): {data_frame['content'][i]}")
    print("-"*15)
```

```python
content_series = data_frame["content"].fillna("")
content_len = content_series.str.len()

print("Shape:", data_frame.shape)
print("Missing content:", data_frame["content"].isna().sum())
print("Empty/blank content:", content_series.str.strip().eq("").sum())
print("Duplicate content:", data_frame["content"].duplicated().sum())
print("Mean chars:", round(content_len.mean(), 2))
print("Median chars:", content_len.median())
print("Min chars:", content_len.min())
print("Max chars:", content_len.max())
print('Contains "Czytaj także":', content_series.str.contains("Czytaj także", case=False, regex=False).sum())
print('Contains "#wieszwiecej":', content_series.str.contains("#wieszwiecej", case=False, regex=False).sum())
```

```python
sample_df = data_frame_full.sample(n=3000, random_state=48)
sample_df = sample_df.reset_index(drop=True)
sample_df = sample_df[['title', 'content']]
print(sample_df.shape)
display(sample_df.head())
```
