```python
import pandas as pd
import numpy as np
from dotenv import load_dotenv
import os

load_dotenv("../.env")
ROOT = os.getenv("PROJECT_ROOT")
```

```python
df = pd.read_csv(f"{ROOT}/data/dev/sygnaly_dev_sample_3000.csv")
display(df.head())
print(df.columns)
print(df.shape)
```

```python
print(df['content'].isnull().sum())
```

```python
df.dropna(subset=['content'], inplace=True)
print(df.shape)
```

```python
df.drop(df[df['content'].str.strip() == ""].index, inplace=True)
print(df.shape)
```

```python
print(df['content'].duplicated().sum())
```

```python
df.drop_duplicates(subset=['content'], inplace=True)
print(df.shape)
```

```python
print(df['content'].str.contains("#wieszwiecej").sum())         # know more
print(df['content'].str.contains("Czytaj także").sum())         # Read also
print(df['content'].str.contains("Polub nas").sum())            # Like us
```

```python
df['content'] = df['content'].str.replace(
    f"#wieszwiecej|Czytaj także:?|Polub nas",
    " ",
    regex=True,
    case=False
)
```

```python
print(f'Mean of length: {df["content"].str.len().mean()}')
print(f'Max of length: {df["content"].str.len().max()}')
print(f'Min of length: {df["content"].str.len().min()}')
```

```python
threshold = 100
df = df[df['content'].str.len() > threshold]
```

```python
# end of cleaning
print(df.shape)
```

```python
df.to_csv(f"{ROOT}/data/dev/sygnaly_dev_clean_3000.csv", index=False)
```
